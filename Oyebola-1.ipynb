```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import math
```

```python
volt = np.arange(-157.95,0,0.05)
```

```python
volt
```

```python
freq = [1/(2*math.pi*math.sqrt(5e-9*1.3637e-3*3.716e-9*math.sqrt(1-v/0.6))) for v in volt]
```

```python
fig, ax = plt.subplots()
ax.plot(volt, freq)

ax.set(xlabel='voltage (V)', ylabel='Frequency (Hz)')

ax.grid()
plt.show()
```
